## Attrition
Let's See The Attrition Data

```python
attrition = raw_data[['EmployeeNumber','Attrition']]
attrition.columns = ['Employees','Attrition']
attrition_total = attrition.groupby(attrition['Attrition']).count()
attrition_percentage = attrition_total/attrition_total.sum()*100
```

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )
    
    
    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Attrition"], loc=legend_)
    # Labels
    plt.ylabel("Employee base (%)")
    plt.show()
    
def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )    
                   
```

```python
attrition_percentage
```

```python
plot_stacked_bars(attrition_percentage.transpose(),"Attrition Status",(5,5),legend_="lower right")
```

## Turnover Analysis

### Department

```python
department = raw_data[['EmployeeNumber', 'Attrition', 'Department']]
department = department.groupby([department['Department'], department['Attrition']])['EmployeeNumber'].count().unstack(level=1).fillna(0)
department_attrition = (department.div(department.sum(axis=1), axis=0) * 100).sort_values(by='Yes', ascending=False)
```

```python
plot_stacked_bars(department_attrition,'Department',rot_=30)
```

### Job Role

```python
role = raw_data[['EmployeeNumber', 'Attrition', 'JobRole']]
role = role.groupby([role['JobRole'], role['Attrition']])['EmployeeNumber'].count().unstack(level=1).fillna(0)
role_attrition = (role.div(role.sum(axis=1), axis=0) * 100).sort_values(by='Yes', ascending=False)
```

```python
# Plot stacked bars
ax = role_attrition.plot(
    kind="bar",
    stacked=True,
    figsize=(18, 10),
    rot=30,
    title="Attrition by Job Role"
)

# Annotate bars
for p in ax.patches:
    value = str(round(p.get_height(), 1))
    if value == '0.0':
        continue
    ax.annotate(
        value,
        ((p.get_x() + p.get_width() / 2) * 0.99 - 0.05, (p.get_y() + p.get_height() / 2) * 0.99),
        color="white",
        size=13
    )

# Rename legend
plt.legend(["Retention", "Attrition"], loc="lower right")

# Labels
plt.ylabel("Employees base (%)")

# Show the plot
plt.show()
```

### Business Travel

```python
# Assuming your DataFrame is named 'raw_data'
# Extract relevant columns
attrition_df = raw_data[['BusinessTravel', 'Attrition', 'Gender']]

# Filter the DataFrame for Attrition 'Yes'
attrition_yes_df = attrition_df[attrition_df['Attrition'] == 'Yes']

# Create a DataFrame for counts
attrition_counts = attrition_yes_df.groupby(['BusinessTravel', 'Gender']).size().unstack(fill_value=0).reset_index()

# Plotting using Seaborn
plt.figure(figsize=(10, 6))

# Create a bar plot
sns.barplot(x='BusinessTravel', y='Female', data=attrition_counts, color='orange', label='Female')
sns.barplot(x='BusinessTravel', y='Male', data=attrition_counts, color='blue', label='Male', bottom=attrition_counts['Female'])

# Customize the plot
plt.title('Attrition Counts (Yes) by BusinessTravel and Gender')
plt.xlabel('BusinessTravel')
plt.ylabel('Attrition Counts (Yes)')
plt.legend(title='Gender', loc='upper right', bbox_to_anchor=(1.15, 1))

# Show the plot
plt.show()
```

### Years In Current Role 

```python
# Assuming your DataFrame is named 'raw_data'
# Extract relevant columns
data = raw_data[['YearsInCurrentRole', 'Attrition', 'JobRole']]

# Define the bins for YearsInCurrentRole
bins = [0, 5, 10, 15, 20]  # Added 1 to include the maximum value

# Create a new column 'YearsInCurrentRoleGroup' based on the bins
data['YearsInCurrentRoleGroup'] = pd.cut(data['YearsInCurrentRole'], bins=bins, labels=False, right=False)

# Count the occurrences of Attrition for each group
attrition_counts = data[data['Attrition'] == 'Yes'].groupby(['YearsInCurrentRoleGroup', 'JobRole']).size().unstack(fill_value=0).reset_index()

# Plotting using Seaborn
plt.figure(figsize=(12, 8))

# Create a dictionary to store the bottom values for each JobRole
bottom_dict = {}

# Iterate through each JobRole and plot stacked bars
for job_role in attrition_counts.columns[1:]:
    plt.bar(attrition_counts['YearsInCurrentRoleGroup'], attrition_counts[job_role], label=job_role, bottom=bottom_dict.get(job_role, 0))
    
    # Update the bottom value for the next iteration
    bottom_dict[job_role] = attrition_counts[job_role] if job_role not in bottom_dict else bottom_dict[job_role] + attrition_counts[job_role]

# Customize the plot
plt.title('Attrition Counts by YearsInCurrentRole and JobRole')
plt.xlabel('Years In Current Role Group')
plt.ylabel('Attrition Counts')
plt.legend(title='JobRole', loc='upper right', bbox_to_anchor=(1.15, 1))

# Manually set x-axis ticks for the specific years
plt.xticks(ticks=[0, 1, 2, 3], labels=['0-5', '5-10', '10-15', '15-20'])

# Show the plot
plt.show()
```

### Age

```python
bins = [18, 25, 35, 45, 55, 60]
labels = ['18-25', '26-35', '36-45', '46-55', '56-60']

# Create ordered categorical variable
raw_data['AgeGroup'] = pd.cut(raw_data['Age'], bins=bins, labels=labels, right=False, ordered=False)

# Group by AgeGroup and Attrition
age_data = raw_data.groupby(['AgeGroup', 'Attrition']).size().unstack(fill_value=0)

# Calculate attrition percentage within each age group
age_data_percentage = (age_data.div(age_data.sum(axis=1), axis=0) * 100).sort_values(by='Yes', ascending=False)

# Manually order the index
age_data_percentage = age_data_percentage.reindex(labels)

# Plot stacked bars
ax = age_data_percentage.plot(
    kind="bar",
    stacked=True,
    figsize=(18, 10),
    rot=30,
    title="Attrition by Age Groups"
)

# Annotate bars
for p in ax.patches:
    value = str(round(p.get_height(), 1))
    if value == '0.0':
        continue
    ax.annotate(
        value,
        ((p.get_x() + p.get_width() / 2) * 0.99 - 0.05, (p.get_y() + p.get_height() / 2) * 0.99),
        color="white",
        size=13
    )

# Rename legend
plt.legend(["Retention", "Churn"], loc="lower right")

# Labels
plt.ylabel("Company base (%)")

# Show the plot
plt.show()
```

### Attrition By age & Gender

```python
age_gender_attrition = raw_data[['AgeGroup', 'Gender', 'Attrition']]

# Filter for only those with Attrition 'Yes'
attrition_yes = age_gender_attrition[age_gender_attrition['Attrition'] == 'Yes']

# Group by AgeGroup and Gender to get the count
attrition_count_by_age_gender = attrition_yes.groupby(['AgeGroup', 'Gender']).size().unstack(fill_value=0)

# Calculate percentages
attrition_percentage_by_age_gender = (attrition_count_by_age_gender.div(attrition_count_by_age_gender.sum(axis=1), axis=0) * 100)

# Plot a grouped bar graph
ax = attrition_percentage_by_age_gender.plot(kind='bar', width=0.35, position=1.5, rot=30)
plt.title('Attrition Percentage by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Percentage')

legend = plt.legend(title='Gender', loc='lower right', fontsize='small')
legend.set_bbox_to_anchor((1.05, 0.5))  # Adjust legend position
legend.get_title().set_fontsize('10')  # Adjust title font size

# Add annotations to the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.1f}%', (x + width/2, y + height), ha='center', va='bottom')

# Show the plot
plt.show()
```
